### Preprocess Notes for NN
#### Store notes and classes 

```python
words = []
outcome = []
for index, row in results.iterrows():
    words.append(word_tokenize(row['text']))
    outcome.append(row['category'])
```

##### Load stopwords and punctuation

```python
nltk.download("stopwords")
stpwrds = stopwords.words('english') + list(string.punctuation)
```

##### Stem words, Remove stopwords and punctuation

```python
# Decided not to use Stemmer for this use case
#stemmer = PorterStemmer()
#lwords = [[stemmer.stem(nw) for nw in w if nw not in stpwrds] for w in words]

twords = []
twords = [[nw for nw in w if nw not in stpwrds] for w in words]
```

#### Identify bigrams in notes

```python
lower_bigram = Phraser(Phrases(twords, min_count=50, threshold=64))
lower_bigram.phrasegrams
```

#### Swap out work pairs for bigrams in notes

```python
clean_words = []
for s in twords:
    clean_words.append(lower_bigram[s])
```

#### Run Word2Vec
(will use Keras embedding first and Word2Vec embedding later. However, have included this section here for later reference. This is not used in futuer sections)

```python
model = Word2Vec(sentences=clean_words, size=64, sg=1, window=50, min_count=40, seed=42, workers=8)
model.save('clean_notes_model.w2v')
model.wv.vocab
```

### Tokenizer text for NN (using Keras embedding) - will use Word2Vec later

#### Flatten all word lists to generate word indices for NN

```python
def flattern(A):
    rt = []
    for i in A:
        if isinstance(i,list): rt.extend(flattern(i))
        else: rt.append(i)
    return rt

fclean_words = flattern(clean_words)
```

#### Isolate Unique Words

```python
uclean_words = []
[uclean_words.append(w) for w in fclean_words if w not in uclean_words]
```

#### Generate word indices

```python
tokenizer = Tokenizer(filters='!"#$%&()*+,-./:;<=>?@[\\]^`{|}~\t\n', oov_token="UKN")
tokenizer.fit_on_texts(uclean_words)
```

#### Use trained indices to convert notes to indices for NN (X Variable)

```python
text_tokens = []
tokens = []
for w in clean_words:
   text_tokens.append(' '.join(w))

X = tokenizer.texts_to_matrix(text_tokens) # dim = number db records x number of unique words
```

#### Create Y Variable

```python
uY = np.unique(np.array(outcome))
Y = np.zeros(shape=(X.shape[0],len(uY)))
Ynum = np.zeros(len(uY))
idx = 0
for s in uY:
    chk = np.array(outcome)==s
    Y[chk,idx]=1
    Ynum[idx] = sum(chk)
    idx += 1
```

#### Create test and training datasets (75% training / 25% testing)

```python
idx = pd.DataFrame(list(range(X.shape[0])))

valid_prcnt = .50
valid_idx = np.random.choice(idx.index, size=round(valid_prcnt*X.shape[0]), replace=False, p=None)
train_idx = ~idx.index.isin(valid_idx)
valid_idx = idx.index.isin(valid_idx)

x_train = X[train_idx,]
y_train = Y[train_idx]

x_valid = X[valid_idx,]
y_valid = Y[valid_idx]

print(x_train.shape)
print(x_valid.shape)
```

## Everything beyong this point is still in draft form and being developed
#### Set hyperparameters

```python
# output directory name:
output_dir = 'model_output/dense' 

# training parameters:
epochs = 20
batch_size = 200

# preprocessing: 
max_note_length = 1000 # Truncate each note to 1000 words
#max_note_length = len(tokenizer.word_index) + 1
pad_type = trunc_type = 'post' # Pad the front so note has 1000 words 
drop_embed = 0.2

# neural network architecture: 
n_dense = 64 # Number of dense layer nodes
n_out = len(uY)
dropout = 0.3 # Fraction of nodes to drop out during training
```

#### Preprocess data

```python
x_train = pad_sequences(x_train, maxlen=max_note_length, padding=pad_type, truncating=trunc_type, value=0)
x_valid = pad_sequences(x_valid, maxlen=max_note_length, padding=pad_type, truncating=trunc_type, value=0)
```

# Basic Dense NN with Textual Data

```python
model = Sequential()
model.add(Dense(n_dense, input_dim=max_note_length))
model.add(Dense(n_dense, activation='relu'))
model.add(Dropout(dropout))
model.add(Dense(n_out, activation='softmax'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
modelcheckpoint = ModelCheckpoint(filepath=output_dir+"/weights.{epoch:02d}.hdf5")
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
model.fit(x_train,y=y_train, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(x_valid, y_valid), callbacks=[modelcheckpoint])
```

#### Evaluate the model

```python
model.load_weights(output_dir+"/weights.19.hdf5") # zero-indexed
```

```python
y_hat = model.predict_proba(x_valid)
```
